```python
from PIL import Image
import rioxarray as rxr
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
xr_pop = rxr.open_rasterio(r'Preprocessed_datasets\tuv_rastpop2020rps_100m.tif',masked=True)
```

```python
xr_nanumea = np.squeeze(np.array(xr_pop.sel(x=slice(176.040783,176.153828),y=slice(-5.623541,-5.712932))))
```

```python
# Pick up sections of shoreline which are close to the buildings
df_shoreline_segments = pd.read_csv('updated_shoreline_proxies.csv')

shoreline_with_pop_dict = {}

dim = 0.0007

for index,row in df_shoreline_segments.iterrows():
    row['pop'] = np.nanmean(xr_pop.sel(x=slice(row.x-dim,row.x+dim),y=slice(row.y+dim,row.y-dim)))
    
    shoreline_with_pop_dict.update({
        index:row
    })
```

```python
df_shoreline_with_pop = pd.DataFrame.from_dict(shoreline_with_pop_dict).T
```

```python
df_shoreline_with_pop['pop_binary'] = [1 if x>=0 else 0 for x in df_shoreline_with_pop['pop']]
```

```python
x = np.array(xr_pop.sel(x=slice(176.040783,176.153828),y=slice(-5.623541,-5.712932)).x)
y = np.array(xr_pop.sel(x=slice(176.040783,176.153828),y=slice(-5.623541,-5.712932)).y)
```

```python
# Check the output in a plot
fig = plt.figure(figsize=(10,10))

ax1 = plt.subplot2grid((1,1),(0,0))


output = ax1.contourf(list(x),list(y),xr_nanumea)

plt.colorbar(output)

ax1.scatter(df_shoreline_with_pop.x,df_shoreline_with_pop.y,c=df_shoreline_with_pop['pop_binary'])

ax1.set_xlim(176.10,176.14)
ax1.set_ylim(np.min(list(y)),-5.65)
```

```python
df_shoreline_with_pop.drop('pop',axis=1,inplace=True)

df_shoreline_with_pop.to_csv('updated_shoreline_proxies_withpop.csv')
```
